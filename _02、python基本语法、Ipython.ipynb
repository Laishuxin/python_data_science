## 鸭子函数

### 判断对象是否可以迭代

```python
def isiterable(obj):
    try:
        iter(obj)  # 如果不可以迭代则会报错，可以迭代则返回一个迭代器
        return True
    except TypeError:
        return False
```

```python
isiterable('string is iterable')
```

## 日期和时间




```python
from datetime import datetime, date, time
dt = datetime(2011, 10, 21, 23, 30, 21)  # 实例化一个时间对象
print(dt.day)
print(dt.minute)
```

```python
# 夺取年月日
# 获取时分秒
print(dt.date())
print(dt.time())
```

```python
# `strftime` 格式化datime
dt2 = dt.strftime('%Y=%m=%d  %H:%M:%S')
print(dt2)
```

```python
# 将字符串转换成datetime对象
# datetime.strptime('str', 'format')
datetime.strptime('2019-01-01 12:21', '%Y-%m-%d %H:%M')
```

```python
# 计算时间差delta

dt1 = datetime(2013,1,2,21,43,12)
dt2 = datetime(2014,1,3,21,43,12)
delta = dt2 - dt1
print(delta)
```
